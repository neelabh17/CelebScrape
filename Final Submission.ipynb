# Getting Personality Traits list
Scrape the web for a personality trait list

```python
#constructin a list of traits
driver=webdriver.Chrome("chromedriver")
driver.get("https://www.vocabulary.com/lists/33665")

content = driver.page_source
soup = BeautifulSoup(content)

ad=(soup.findAll("a", attrs={'class':'word dynamictext'}))
traits=[]
for a in ad:
    traits+=a.contents

# close the webpage
driver.close()
print(traits)
```

# Get names of celebs
Getting names of around top 200 indian celebreties


```python
# get source from webpage
driver=webdriver.Chrome("chromedriver")
driver.get("https://www.indiaforums.com/person/list")
time.sleep(2.5)

numpages=1
celebList=[]

for i in range(numpages):
    oh=0
    nh=driver.execute_script("return document.body.scrollHeight")
    while(nh!=oh):
        time.sleep(0.5)
        oh=nh
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(0.5)
        nh=driver.execute_script("return document.body.scrollHeight")

    #extract data    
    content = driver.page_source
    soup = BeautifulSoup(content)  
    
    #extracting names
    res=soup.findAll("div", attrs={'class':'celebThumb'})
    for a in res:
        celebList.append(a.img.get("alt"))
    
    #going to next page
    nextbutton=driver.find_elements_by_xpath("//*[@id='bottomNext']")[0]
    nextbutton.click()
    
    
print(celebList)
print("Number of Celebs : {}".format(len(celebList)))

#close webpage    
driver.close()


```

```python
def getWords(sentence):
    def rm(s):
        l=len(s)
        i=0
        alt=[]
        while(i<l):
            if not (len(s[i])<=3 or "[" in s[i] or "]" in s[i]) :
                alt+=s[i]
            i+=1
        return alt

    s=sentence.split(" ")
    i=0
    ns=[]
    while(i<len(s)):
        d=s[i]
        d=d.split(".")
        ns+=d
        i+=1
    ns
    ns=rm(ns)
    print(ns)
    s=ns.copy()

    ns=[]
    i=0
    while(i<len(s)):
        d=s[i]
        d=d.split(",")
    #     print(i,d)
        ns+=d
        i+=1
    nm=rm(ns)
    return nm


```
